# **Milage Prediction**

-------------

## **Objective**

Predict milage of a vehicle using linear regression model

## **Data Source**

https://github.com/YBI-Foundation/Dataset

## **Import Library**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## **Import Data**

```python
df = pd.read_csv('https://github.com/YBI-Foundation/Dataset/raw/main/MPG.csv')
```

## **Describe Data**

```python
df.head()
```

```python
df.nunique()
```

## **Data Preprocessing**

```python
df.info()
```

```python
df.describe()
```

```python
print(df.corr())
```

```python
df = df.dropna()
```

```python
df.info()
```

```python
df.columns
```

## **Data Visualization**

```python
sns.pairplot(df, x_vars =['displacement', 'horsepower', 'weight',
       'acceleration','mpg'], y_vars = ['mpg']);
```

```python
sns.regplot(x = 'displacement', y = 'mpg', data = df);
```

## **Define Target Variable (y) and Feature Variables (X)**

```python
y = df['mpg']
```

```python
y.shape
```

```python
X = df.drop(['mpg', 'cylinders','model_year', 'origin', 'name'], axis = 1)
```

```python
X.shape
```

```python
X
```

# **Scaling Data**

```python
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()

X = ss.fit_transform(X)

X
```

```python
pd.DataFrame(X).describe()
```

## **Train Test Split**

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,train_size = 0.7, random_state =2529)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

## **Modeling**

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()

lr.fit(X_train, y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

## **Prediction**

```python
y_pred = lr.predict(X_test)
```

```python
y_pred
```

```python
plt.scatter(y_test, y_pred, color='blue')

# Add a diagonal line to represent perfect predictions
plt.plot([np.min(y_test), np.max(y_test)], [np.min(y_test), np.max(y_test)], color='red')

# Add labels and a title to the plot
plt.xlabel('Actual mileage')
plt.ylabel('Predicted mileage')
plt.title('Linear regression for mileage prediction')

# Display the plot
plt.show()
```

## **Model Evaluation**

```python
from sklearn.metrics import mean_absolute_error, mean_absolute_percentage_error, r2_score  
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
mean_absolute_percentage_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```
